````python
# 첫 풀이 (시간 초과)
# 합계: 30.0 / 100.0

# [3, 1, 2, 3] 번 폰켓몬

from itertools import combinations
from collections import Counter

def solution(nums):
    answer = 1 # max 값 구해야
    all_cases = combinations(nums, len(nums)//2)
    
    for case in all_cases:
        #print(case)
        
        c = Counter(case)
        #print(c)
        #print(len(c))
        
        if answer < len(c):
            answer = len(c) # 값 업데이트
    
    return answer
````

- 징검다리

````python
# refer (성공)
# https://western-sky.tistory.com/135

def solution(distance, rocks, need_n):
    answer = 0
    
    rocks.sort()
    rocks.append(distance)
    
    s = 0
    e = distance
    
    while s <= e:
        mid = (s+e) // 2
        
        last_rock_location = 0
        remove_cnt = 0
        
        for now_location in rocks:
            if (now_location - last_rock_location) < mid:
                # 삭제 가능
                remove_cnt += 1
                
            else:
                # 삭제 불가능
                last_rock_location = now_location
                
                
        # 탐색 후 다음 이분탐색 앞쪽/뒤쪽 결정
        if remove_cnt > need_n:
            # "제거가능한 지점 사이의 거리의 최솟값"을 더 줄일 수 있음
            # 즉, 앞쪽 탐색
            e = (mid - 1)
            ###### (여기에 놓으면 안 됨) answer = mid
            
            
        else:
            # 아직 제거하는 바위돌의 개수 부족 -> "제거가능한 지점 사이의 거리의 최솟값"을 더 늘려야 함
            s = (mid + 1)
            answer = mid
    
    return answer # 바위를 n개 제거한 뒤 각 지점 사이의 거리의 최솟값 중에 가장 큰 값을 return
````

````python
# 노드 번호는 0이 아닌 1부터 시작


from collections import deque, Counter

def solution(n, vertex):
    answer = 0
    depth_of_each_node = [-1] * (n+1)
    depth_of_each_node[1] = 0
    
    
    a = [[] for _ in range(n+1)]
    
    for (u,v) in vertex:
        a[u].append(v)
        a[v].append(u)
        
        
    # 큐 1개 사용해 탐색 시작
    # 항상 1번 노드부터 가장 멀리 떨어진 정도(=depth) 탐색 !
    q = deque()
    q.append([1, 0])  # 노드 번호 idx, 해당 노드의 depth
    
    while q:
        node, depth = q.popleft()
        
        for y in a[node]:
            if depth_of_each_node[y] == -1: # (중요 !) 첫 방문일 때만 수행 (1번 노드로부터 가장 멀리 떨어진 정도를 구하는 것이므로)
                q.append([y, depth+1])
                depth_of_each_node[y] = depth+1
    
    
    ###print(depth_of_each_node)
   

    # 정답 구하기
    max_depth = max(depth_of_each_node)
    
    if max_depth == 0 or max_depth == -1:
        return -1
    
    
    c = Counter(depth_of_each_node)
    return c[max_depth]
````

````python
# 지난 풀이 (백업)

def solution(grid):
    answer = []
    
    r = len(grid)
    c = len(grid[0])
    
    visited = [[ [False for _ in range(4)] for _ in range(c)] for _ in range(r)] # r행 c열에서 d방향으로 나간 적 있는지 기록하는 3차원 행렬
    
    # for row in visited:
    #     print(row)
        
    def move(i, j, d):
        if d == 0: # 위
            i = (i-1+r) % r
            
        if d == 1:  # 오른쪽
            j = (j+1) % c
        
        if d == 2:  # 아래
            i = (i+1) % r
        
        if d == 3:  # 왼쪽
            j = (j-1+c) % c
                 
        return (i, j)

        
    ## d = {0: 위, 1: 오른쪽, 2: 아래, 3: 왼쪽}     (시계 방향으로 지정)
    def direction(ni, nj, d):
        if grid[ni][nj] == "S": # 직진
            return d
        
        elif grid[ni][nj] == "L": # 좌회전 (-1 방향)
            return ((d-1)+4) % 4
        
        elif grid[ni][nj] == "R": # 우회전 (+1 방향)
            return (d+1) % 4
        
    
    # 완전 탐색 - 총 r개 행, c개 열에서 모든 4가지 방향으로 시작해보면서 경우의 수 찾기
    for i in range(r):
        for j in range(c):
                
            for d in range(4):
                if visited[i][j][d] == False: # 다음 시작점
                    now_x, now_y, now_d = i, j, d
                    cnt = 0
                    ###print('시작', i, j, d)
                    
                    while visited[now_x][now_y][now_d] is False:
                        # 사이클 방문 수행
                        visited[now_x][now_y][now_d] = True
                        next_x, next_y = move(now_x, now_y, now_d)
                        next_d = direction(next_x, next_y, now_d) # 다음 좌표에 의해 방향이 바뀜
                        cnt += 1 # 사이클의 크기 +1 증가
                        
                        now_x, now_y, now_d = next_x, next_y, next_d
                        ###print(now_x, now_y, now_d)
                    
                    
                    ###print(cnt)
                    answer.append(cnt)
        
    answer.sort() # 주어진 격자를 통해 만들어지는 빛의 경로 사이클의 모든 길이들을 배열에 담아 오름차순으로 정렬하여 return
    return answer
````

````python
# 성공

from collections import deque

def solution(numbers, target):
    answer = 0
    # node 들의 숫자번호 = numbers 리스트의 각 index 번호
    
    q = deque([ [-1, 0] ]) # [시작 인덱스, 현재까지의 sum 값]
    
    while q:
        pre_x, pre_sum = q.popleft()
        
        now_x = pre_x + 1
        if now_x == len(numbers): # 1) 그만
            if pre_sum == target:
                answer += 1
            #######break
            
        else: #) 다음 큐 이어서 생성
            q.append([now_x, (pre_sum + numbers[now_x])]) # 1) 더해보기
            q.append([now_x, (pre_sum - numbers[now_x])]) # 2) 빼보기
    
    return answer
````

````python
# 첫 시도 (절반만 성공)
# 정확성: 50.0 / 100.0


from collections import deque

def solution(N, road, K):
    answer = 0
    # 노드의 번호 : 1 ~ N

    # 1. 인접 행렬 준비
    a = [ [[] for _ in range(N+1)] for _ in range(N+1) ]
    # for row in a:
    #     print(row)
        
    for (u, v, time) in road:
        a[u][v].append(time)
        a[v][u].append(time)
        
    for i in range(1, N+1):
        for j in range(1, N+1):
            a[i][j].sort() # 시간이 적게 걸리는 순으로 오름차순 정렬 !
            
            
    
    # 2. 탐색 시작
    check = [False] * (N+1)
    time_lt = [-1] * (N+1)
    
    q = deque()
    # 항상 1번 노드부터 시작
    q.append([1, 0]) # [노드 번호, 걸리는 시간]
    check[1] = True
    time_lt[1] = 0 # 1번 노드는 무조건 0초 걸림
    
    while q:
        x, pre_sum = q.popleft()
        
        for y in range(1, N+1):
            times_of_cols = a[x][y]

            if times_of_cols == []:
                continue
                
            if check[y] == False:
                q.append([y, (pre_sum + times_of_cols[0])])
                check[y] = True
                
                if time_lt[y] == -1:
                    time_lt[y] = (pre_sum + times_of_cols[0])
                else:
                    time_lt[y] = min((pre_sum + times_of_cols[0]), time_lt[y]) # MIN 값으로 새로 구한 배달 소요시간 업데이트
                
                
    # print(check)
    # print(time_lt)
    
    
        
        
    # 3. 정답 구하기
    for spend_time in time_lt[1:]:
        if spend_time <= K:
            answer += 1
    

    return answer
````

````python
# 지난 풀이 (백업)

def solution(n):
    answer = []
    
    a = [[0]*n for _ in range(n)] # NxN 행렬
    
    k = 0 # 아래 방향에서 시작
    row = 0
    col = 0
    a[row][col] = 1
    
    number = 2
    
    for _ in range(n*n):
        if number > n*(n+1)//2:
            break
            
        
        if k==0:
            if row+1 >= n or a[row+1][col] != 0:
                k = (k+1)%3
                continue
            
            row += 1
            a[row][col] = number
            number += 1
            continue
            
        if k==1:
            if col+1 >= n or a[row][col+1] != 0:
                k = (k+1)%3
                continue
                
            col += 1
            a[row][col] = number
            number += 1
            continue
            
            
        if k==2:
            if a[row-1][col-1] != 0:
                k = (k+1)%3
                continue
                
            row -= 1
            col -= 1
            a[row][col] = number
            number += 1
            continue

    #for row in a:
    #    print(row)
    
    for row_cnt in range(1, n+1):
        answer.extend(a[row_cnt-1][:row_cnt])
    
    
    return answer
````
